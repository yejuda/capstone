```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from konlpy.tag import Okt

driver = webdriver.Chrome('C:/chromedriver_win32/chromedriver.exe')

driver.get('https://www.mangoplate.com/restaurants/9Gi6030AUV')

body = driver.find_element(By.TAG_NAME, 'body')

num_of_pagedowns = 20

while num_of_pagedowns:

    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(1.5)
    num_of_pagedowns -= 1

    try:
        driver.find_element(By.XPATH, "//button[@class='btn-reivews-more']")\
            .click()
    except:
        break
    
html = driver.page_source
result = BeautifulSoup(html,'html.parser')
    
body = result.find("body")
div = body.select('div.review_wraper > span')


for content in div:
    txt = content.get_text().strip()
    with open('matzip.txt','a',encoding='UTF-8') as f:
        f.write(txt+'\n')

driver.get('https://www.siksinhot.com/P/5998#')

body = driver.find_element(By.TAG_NAME, 'body')

num_of_pagedowns = 12

while num_of_pagedowns:

    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(1.5)
    num_of_pagedowns -= 1

    try:
        driver.find_element(By.XPATH, "//a[@class='btn_sMore']")\
            .click()
    except:
        break
    
html = driver.page_source
result = BeautifulSoup(html,'html.parser')
    
body = result.find("body")
div = body.select("div.score_story > p")

for ptag in div:
    txt = ptag.get_text().strip()
    with open('matzip.txt','a',encoding='UTF-8') as f:
        f.write(txt+'\n')
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from konlpy.tag import Okt

driver = webdriver.Chrome('C:/chromedriver_win32/chromedriver.exe')
driver.get('https://www.siksinhot.com/P/5998#')

body = driver.find_element(By.TAG_NAME, 'body')

num_of_pagedowns = 12

while num_of_pagedowns:
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(1.5)
    num_of_pagedowns -= 1
    
    try:
        more_button = driver.find_element(By.XPATH, "//a[@class='btn_sMore']")
        if more_button.is_displayed():
            more_button.click()
            time.sleep(1.5)
    except:
        break

html = driver.page_source
result = BeautifulSoup(html,'html.parser')
    
body = result.find("body")
div = body.select("div.score_story > p")

for ptag in div:
    txt = ptag.get_text().strip()
    with open('matzip.txt','a',encoding='UTF-8') as f:
        f.write(txt+'\n')
        
driver.quit()
```
